Now based on this equation : $mgs\sin(\phi) = I \ddot{\phi}$, we need to calculate the angular acceleration first and then derive the moment of inertia.
To derive the angular acceleration, we need to differentiate the graph of angular velocity:


```python
# @title Default title text
import pandas as pd
import matplotlib.pyplot as plt


url = 'https://github.com/delaramroozbehani/lab1/raw/main/Gyroscope%20rotation%20rate%202023-11-13_15-40-11%20x.csv'
df = pd.read_csv(url)


df['Angular Acceleration (rad/s^2)'] = df['Gyroscope x (rad/s)'].diff()


plt.figure(figsize=(10, 6))
plt.plot(df['Time (s)'], df['Gyroscope x (rad/s)'], label='Angular Velocity')
plt.plot(df['Time (s)'], df['Angular Acceleration (rad/s^2)'], label='Angular Acceleration')
plt.title('Angular Velocity and Acceleration Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Angular Values')
plt.legend()
plt.grid(True)
plt.show()
```

Then by calculating the mean angular acceleration, considering the angle at which the toppling happens, sin(Ï•)= sin(Ï€/2) = 1
##### This gives: $I = \frac{mgs}{\ddot{\phi}}
$
Where m = mass of the phone = 157g,
##### $g = 9.81 m/s^2$
$s_x = 0.15 m (length-of -phone) / 2 = 0.07 m$


```python
# @title Default title text
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


url = 'https://github.com/delaramroozbehani/lab1/raw/main/Gyroscope%20rotation%20rate%202023-11-13_15-40-11%20x.csv'
df = pd.read_csv(url)


mass = 0.157


g = 9.81


s = 0.075


df['Angular Acceleration (rad/s^2)'] = df['Gyroscope x (rad/s)'].diff()


df = df.dropna()


angular_acceleration_column = 'Angular Acceleration (rad/s^2)'
angular_acceleration = df[angular_acceleration_column]


moment_of_inertia = round((mass * g * s) / angular_acceleration.mean(), 3)



print("Estimated Moment of Inertia:", moment_of_inertia)
```

# Task 2, part c:
since we have already sketched the angular velocity about x and y axis, in this section we're only calculating moment of inertia based on $\dot\phi^2$ and $cos^2(Ï•/2)$

##### moment of inertia about x axis:

$\dot{\phi}^2 = 2E_0 /I - 4\omega_0^2 \cos^2(\phi/2)$

```python
# @title Default title text
import pandas as pd
import numpy as np

# Load the CSV file into a DataFrame
url = 'https://github.com/delaramroozbehani/lab1/raw/main/Gyroscope%20rotation%20rate%202023-11-13_15-40-11%20x.csv'
df = pd.read_csv(url)

# Extract relevant columns from the DataFrame
phi_dot = df['Gyroscope x (rad/s)']

# Constants
m = 0.157  # Your mass value
g = 9.81   # Your gravitational acceleration value
s = 0.075  # Your s value

# Calculate the mean angular velocity
mean_phi_dot = np.mean(phi_dot)

# Specify the angle of interest
phi_interest = np.pi / 2

# Calculate moment of inertia at the specified angle using the mean angular velocity
I_at_phi_interest = round((4 * m * g * s * np.cos(phi_interest/2)**2) / (mean_phi_dot**2), 4)

# Print the result
print(f'Moment of Inertia at phi = pi/2 using mean angular velocity: {I_at_phi_interest}')
```

Task 2, a : moment of inertia based on : $\dot{\phi}(t) = \frac{\sqrt{mgs}}{I\exp\left(-\sqrt{\frac{mgs}{I}}t + \left(\frac{\phi_0}{4}\right)^2\right)} \exp\left(\sqrt{\frac{mgs}{I}}t\right)$

about x axis:

```python
# @title Default title text
import numpy as np
from scipy.integrate import odeint
from scipy.optimize import minimize
import pandas as pd

data = pd.read_csv('https://github.com/delaramroozbehani/lab1/raw/main/Gyroscope%20rotation%20rate%202023-11-13_15-40-11%20x.csv')
t_data = data['Time (s)'].values
phi_dot_data = data['Gyroscope x (rad/s)'].values

def pendulum_eq(phi, t, m, g, I, phi0, s):
    return (np.sqrt(m * g * s / I) * np.exp(-np.sqrt(m * g * s / I) * t + (phi0 / 4)**2) * np.exp(np.sqrt(m * g * s / I) * t))

def objective(I, *args):
    t_data, phi_dot_data, m, g, phi0, s = args
    phi_numerical = odeint(pendulum_eq, phi0, t_data, args=(m, g, I, phi0, s))
    return np.sum((phi_numerical[:, 0] - phi_dot_data)**2)

m_value = 0.157
g_value = 9.81
phi0_value = np.pi / 2
s_value = 0.075

result = minimize(objective, x0=[1.0], args=(t_data, phi_dot_data, m_value, g_value, phi0_value, s_value))

I_optimized = result.x[0]
print("Optimized value of I:", I_optimized)

```
